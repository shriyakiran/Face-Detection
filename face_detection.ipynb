```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import dlib
```

```python
image=cv2.imread('people1.jpg')
image
```

```python
image.shape
```

```python
image=cv2.resize(image,(800,600))
image.shape
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```python
face_detector=cv2.CascadeClassifier('C:/Users/dshri/Downloads/haarcascade_frontalface_default.xml')
display(face_detector)
```

```python
detections=face_detector.detectMultiScale(image_gray)
display(detections)
```

```python
display(len(detections))
```

```python
for(x,y,w,h) in detections:
    cv2.rectangle(image_gray,(x,y),(x+w,y+h),(0,255,255),4)
plt.imshow(image_gray)
#cv2.waitKey(20000)
#plt.show()
#cv2.destroyAllWindows()
```

```python
image=cv2.imread('group.jpg')
image=cv2.resize(image,(800,600))
plt.imshow(image)
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
detections=face_detector.detectMultiScale(image_gray,scaleFactor=1.09,minNeighbors=9)
display(detections)
```

```python
for(x,y,w,h) in detections:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),4)
plt.imshow(image)
cv2.waitKey(20000)
plt.show()
cv2.destroyAllWindows()
```
